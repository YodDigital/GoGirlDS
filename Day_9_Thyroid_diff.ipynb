## Visualisations and EDA

```python
thyroid['T'].value_counts().plot(kind='bar')
```

```python
thyroid['N'].value_counts().plot(kind='bar')
```

```python
thyroid['M'].value_counts().plot(kind='bar')
```

```python
thyroid['Recurred'].value_counts().plot(kind='bar')
```

```python
thyroid['Age_groups'] = thyroid['Age'].apply(lambda x: 'T' if x < 20 else ('YA' if x < 40 else ('MA' if x<60 else 'OA')))
axe = thyroid['Age_groups'].value_counts().plot(kind='bar')
axe.bar_label(axe.containers[0])
```

```python
# prompt: plot the graph of value counts of the Age_group column based on the recurred column and add the value counts ontop of each bar

import matplotlib.pyplot as plt

df = thyroid.groupby(['Age_groups', 'Recurred']).size().unstack().fillna(0)

ax = df.plot(kind='bar', stacked=False)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.show()


```

```python
df = thyroid.groupby(['Response', 'Recurred']).size().unstack().fillna(0)

ax = df.plot(kind='bar', stacked=False)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.show()
```
