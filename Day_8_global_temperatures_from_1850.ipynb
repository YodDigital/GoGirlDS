# Data Cleaning

```python
global_temp['GlobalTemperatures'] = pd.to_datetime(global_temp['GlobalTemperatures'], format='%Y-%m-%d')
global_temp.head()
```

```python
global_temp_jan = global_temp.loc[global_temp['GlobalTemperatures'].dt.month==1, :]
global_temp_jan.head()
```

```python
global_temp_jan.shape
```

# Visualisation

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Winter

```python
plt.figure(figsize=(10, 6))
# sns.histplot(data=global_temp_jan, x='LandAverageTemperature', bins=30)
sns.lineplot(x='GlobalTemperatures', y=	'LandAverageTemperature', data=global_temp.loc[global_temp['GlobalTemperatures'].dt.month.isin([12,1,2]), :])
plt.title('Distribution of Temperature over the years in winter')
plt.xlabel('Years')
plt.ylabel('Temperatures')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# sns.histplot(data=global_temp_jan, x='LandAverageTemperature', bins=30)
sns.lineplot(x='GlobalTemperatures', y=	'LandAndOceanAverageTemperature', data=global_temp.loc[global_temp['GlobalTemperatures'].dt.month.isin([12,1,2]), :])
plt.title('Distribution of Temperature (land and ocean) over the years in winter')
plt.xlabel('Years')
plt.ylabel('Temperatures')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# sns.histplot(data=global_temp_jan, x='LandAverageTemperature', bins=30)
sns.lineplot(x='GlobalTemperatures', y=	'LandAndOceanAverageTemperatureUncertainty', data=global_temp.loc[global_temp['GlobalTemperatures'].dt.month.isin([12,1,2]), :])
plt.title('Distribution of LandAndOceanAverageTemperatureUncertainty over the years in winter')
plt.xlabel('Years')
plt.ylabel('Temperatures')
plt.grid(True)
plt.show()
```
